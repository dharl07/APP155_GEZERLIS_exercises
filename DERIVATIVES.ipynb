```python
import matplotlib.pyplot as plt
import numpy as np

from decimal import Decimal
from fractions import Fraction
```

```python
# WORKED EXAMPLE 3.8

x = 0.004

def f_func(x):
    f = (1 - np.cos(x))/x**2
    return f

f_prime = (-x*np.sin(x) + 2*np.cos(x) - 2)/x**3

f_double_dev = ((-4*x*np.sin(x) + (x**2)*np.cos(x) - 6*np.cos(x) + 6))/x**4


def forward_diff(f_func, x, h):
    fd = (f_func(x+h) - f_func(x))/h
    return fd

def cen_diff(f_func, x, h):
    cd = (f_func(x+h) - 2*f_func(x) + f_func(x-h))/h**2
    return cd

hs = [10**(-i) for i in range(1,7)]
hs = np.logspace(-7, -1, 200) 
ans = [abs(cen_diff(f_func, x, h) - f_double_dev) for h in hs]

print("analytic second derivative", f_double_dev)

for i in range(len(hs)):
    print(ans[i])
```

```python
fig, ax = plt.subplots()

ax.loglog(hs, ans)
```

```python
# Introducing a new set of points x = 0.004, with the new function

x = 0.004

def f2_func(x):
    f2 = (1/2)*(np.sin(x/2))**2/(x/2)**2
    return f2

def central_diff(f2_func, x, h2):
    cd2 = (f2_func(x+h2) - 2*f2_func(x) + f2_func(x-h2))/h2**2
    return cd2
```

```python
hs2 = [10**(-i) for i in range(1,7)]
hs2 = np.logspace(-7, -1, 200) 
abs_2 = [abs(central_diff(f2_func, x, h2) - f_double_dev) for h2 in hs2]

for i in range(len(hs2)):
    print(abs_2[i])
```

```python
fig, ax = plt.subplots()


ax.loglog(hs2, abs_2)
```

```python
fig, ax = plt.subplots()

ax.set(
    xlabel = "Values of h",
    ylabel = "Absolute Error",
    title = "Absolute Error",
)
ax.loglog(hs, ans, color="red")
ax.loglog(hs2, abs_2, color="black")
```
